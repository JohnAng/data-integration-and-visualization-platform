```python
import polars as pl
from pathlib import Path

BASE_DIR = Path("../data")

pl.Config.set_tbl_cols(30)
pl.Config.set_tbl_formatting("ASCII_MARKDOWN")

def inspect_dataset(folder: str, filename: str, separator: str = ";") -> pl.DataFrame:
    """
    Reads the first 5 rows of a delimited file.
    Applies truncate_ragged_lines to bypass schema mismatch errors caused by unescaped delimiters.
    """
    file_path = BASE_DIR / folder / filename
    df = pl.read_csv(
        file_path,
        n_rows=5,
        separator=separator,
        ignore_errors=True,
        infer_schema_length=0,
        truncate_ragged_lines=True
    )
    return df
```

```python
"""
Preview DBLP Conferences dataset.
"""
df_inproceedings = inspect_dataset("dblp_dataset", "input_inproceedings.csv", separator=";")
print("=== COLUMNS ===")
print(df_inproceedings.columns)
print("\n=== DATA ===")
print(df_inproceedings.head())
```

```python
"""
Extracts the schema for Journal Articles and Journal Rankings.
"""
df_articles = inspect_dataset("dblp_dataset", "input_article.csv", separator=";")
print("=== JOURNAL ARTICLES COLUMNS ===")
print(df_articles.columns)

df_rankings_comma = inspect_dataset("journal_ranking_data_raw", "journal_ranking_data_raw.csv", separator=",")
df_rankings_semi = inspect_dataset("journal_ranking_data_raw", "journal_ranking_data_raw.csv", separator=";")

if len(df_rankings_comma.columns) > len(df_rankings_semi.columns):
    print("\n=== JOURNAL RANKINGS COLUMNS (COMMA SEPARATED) ===")
    print(df_rankings_comma.columns)
else:
    print("\n=== JOURNAL RANKINGS COLUMNS (SEMICOLON SEPARATED) ===")
    print(df_rankings_semi.columns)
```

```python
"""
Extracts Journal titles from DBLP and Kaggle, and inspects iCore26 data.
"""
df_articles = inspect_dataset("dblp_dataset", "input_article.csv", separator=";")
print("=== DBLP JOURNAL TITLES SAMPLE ===")
print(df_articles.select(["journal"]).head())

df_rankings = inspect_dataset("journal_ranking_data_raw", "journal_ranking_data_raw.csv", separator=",")
print("\n=== KAGGLE JOURNAL TITLES SAMPLE ===")
print(df_rankings.select(["Title"]).head())

df_icore = inspect_dataset("icore26_data", "iCore26_KilledColumnsForLoading.csv", separator=",")
print("\n=== ICORE26 COLUMNS ===")
print(df_icore.columns)
print("\n=== ICORE26 DATA SAMPLE ===")
print(df_icore.select(df_icore.columns[:5]).head())
```
